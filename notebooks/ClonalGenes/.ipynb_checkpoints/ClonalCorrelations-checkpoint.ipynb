```python
num_tests = 1000
obs_values_list = []
exact_matches = [] 
for i in range(_df_switchclones.shape[0]):
    obs_vals = _df_switchclones.iloc[i].values
    trials = _df_switchclones.iloc[i].sum()
    rng = np.random.default_rng()
    sim_outcomes = rng.multinomial(trials, ens_switchprobs, size=num_tests)
    counter = 0
    for i in range(num_tests):
        # Test for equality of outcome of sim and obs
        counter += np.array_equal(obs_vals, sim_outcomes[i])
    obs_values_list.append(obs_vals)
    exact_matches.append(counter)
```

```python
## Adapted from https://gist.github.com/ShinNoNoir/4749548
#https://github.com/amirziai/learning/blob/master/statistics/Inter-rater%20agreement%20kappas.ipynb

def fleiss_kappa(ratings, n):
    '''
    Computes the Fleiss' kappa measure for assessing the reliability of 
    agreement between a fixed number n of raters when assigning categorical
    ratings to a number of items.
    
    Args:
        ratings: a list of (item, category)-ratings
        n: number of raters
        k: number of categories
    Returns:
        the Fleiss' kappa score
    
    See also:
        http://en.wikipedia.org/wiki/Fleiss'_kappa
    '''
    items = set()
    categories = set()
    n_ij = {}
    
    for i, c in ratings:
        items.add(i)
        categories.add(c)
        n_ij[(i,c)] = n_ij.get((i,c), 0) + 1
    
    N = len(items)
    
    p_j = dict(((c, sum(n_ij.get((i, c), 0) for i in items) / (1.0 * n * N)) for c in categories))
    P_i = dict(((i, (sum(n_ij.get((i, c), 0) ** 2 for c in categories) - n) / (n * (n - 1.0))) for i in items))

    P_bar = sum(P_i.values()) / (1.0 * N)
    P_e_bar = sum(value ** 2 for value in p_j.values())
    
    kappa = (P_bar - P_e_bar) / (1 - P_e_bar)
    
    return kappa
```
